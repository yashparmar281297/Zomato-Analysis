# Zomatao Dataset Analysis

```python
# First we import the csv file and important libraries. 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
zomato = pd.read_csv("C:/Users/yashp/Downloads/archive (1)/zomato.csv")
```

# Data understanding and preprocessing

```python
# now we explore oru data and do data preprocessing 
zomato.head()
```

```python
zomato.info()
```

```python
nullcounts = zomato.isnull().sum()
print(nullcounts)

len(zomato)
```

```python
percentnullcounts = (nullcounts/len(zomato))*100
print(percentnullcounts)
```

```python
#NOW WE HAVE TO REMOVE NULL VALUES 
zomato= zomato.dropna()
```

```python
nullcounts = zomato.isnull().sum()
print(nullcounts)
```

```python
zomato['approx_cost(for two people)'] = zomato['approx_cost(for two people)'].str.replace(',', '').astype(float).astype(int)
```

```python
zomato.info()
```

```python
# Now we remove the columns which are not neccessary for our analysis .
# the columns are url, reviews_list, menu_list

columns_to_remove=['url','reviews_list','menu_item']
zomato=zomato.drop(columns=columns_to_remove)
print(zomato.head())
```

```python
#now we have to remove /5 from rate because we know that rating is done on scale of 5 
# Replace "NEW" with 0 and 0 is representing new 
# Convert the 'rate' column to string
zomato['rate'] = zomato['rate'].astype(str)
zomato['rate'] = zomato['rate'].replace('NEW', '0')
zomato['rate']= zomato['rate'].str.replace('/5','').str.strip().astype(float) #.str.strip(): After replacing "/5" with an empty string, the .str.strip() function is applied. This function is used to remove any leading or trailing whitespace (spaces, tabs, or other whitespace characters) from each cell in the "rate" column. This step ensures that there are no unwanted spaces around the numeric ratings.
print(zomato['rate'].head())
```

```python
# now we have done with all the data cleaning and preprocessing. so firs we save this data to a csv file and do visualization 
zomato.to_csv('C:/Users/yashp/Downloads/cleaned_zomato_data.csv',index = False)
```

# Visualization 

```python
!pip install wordcloud
from wordcloud import WordCloud
import seaborn as sns
```

```python
# Distribution of ratings 
plt.figure(figsize=(10,6))
sns.histplot(zomato['rate'],bins=20, kde=True)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('frequency') 
```

```python
# Analyze which  dish_liked, cuisines, rest_type comes most number of time using word chart. 
#first we make word chart for dish_liked
all_dishes = ''.join(zomato['dish_liked'].dropna())
wordcloud = WordCloud(width=800,height=500,background_color='white').generate(all_dishes)
plt.figure(figsize=(10,8))
plt.imshow(wordcloud,interpolation= 'bilinear')
plt.title('Word Cloud of Popular Dishes')
plt.axis('off')
plt.show()
```

```python
#Cloud chart for cuisines
all_cuisines=''.join(zomato['cuisines'].dropna())
wordcloud= WordCloud(width=800,height=500,background_color='white').generate(all_cuisines)
plt.figure(figsize=(10,8))
plt.imshow(wordcloud,interpolation='bilinear')
plt.title('Word Cloud of Popular Cuisines')
plt.axis('off')
plt.show()
```

```python
#Which location has how many restaurant
location_counts=zomato['location'].value_counts()  
# if i want only top 20 then we have to add one more line of code
top_location= location_counts.head(20)
plt.figure(figsize=(12,6))
top_location.plot(kind='bar',color='skyblue')
plt.title('Number of Restaurants by Location(top 20)')
plt.xlabel('Location')
plt.ylabel('Number of Restaurants') 
plt.tight_layout()
plt.show()
```

```python
# Most popular cuisines 
cuisines_counts= zomato['cuisines'].str.split(', ').explode().value_counts()
top_cuisines = cuisines_counts.head(10)
plt.figure(figsize=(6,6))
plt.pie(top_cusines, labels= top_cuisines.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Most Popular Cuisines')
```

```python
# Finding is there any relation between numercial values like rate, vote and approx_cost by correlation analysis
numerical_columns= ['rate','votes','approx_cost(for two people)'] 
correlation_matrix = zomato[numerical_columns].corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True,cmap='coolwarm',linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
#How much restaurant allow online delivery or not
online_order_counts= zomato['online_order'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(online_order_counts,labels=online_order_counts.index, autopct='%1.1f%%',startangle=75)
plt.title('Distribution of Restaurant Allowing Onine Delivery')
```

```python
# How much restaurant allowing table booking or not
table_booking_counts= zomato['book_table'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(table_booking_counts,labels=table_booking_counts.index, autopct='%1.1f%%', startangle=75)
plt.title('Distribution of Restaurant Allowing Booking Table')
```

```python
restaurant_counts= zomato['rest_type'].str.split(', ').explode().value_counts()
plt.figure(figsize=(10,8))
restaurant_counts.plot(kind='bar',color='green')
plt.title('Count of Restaurant Type ')
plt.xlabel('Restaurant type')
plt.ylabel('Number of Restaurants') 
plt.tight_layout()
plt.show()
```

```python
# Distribution of approx cost for 2 person
plt.figure(figsize=(10,6))
sns.histplot(zomato['approx_cost(for two people)'],bins=20,color='purple',edgecolor='black',kde=True)
plt.title('Distribution of Approx Cost for 2 Person')
plt.xlabel('Approx Cost for 2(in INR)')
plt.ylabel('Frequncy')
plt.tight_layout()
```
